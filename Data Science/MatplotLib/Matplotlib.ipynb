<font size ='5'>MatplotLib Tutorial</font>

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt or GTK+.

Some of the major Pros of Matplotlib are:
<li>Generally easy to get started for simple plots</li>
<li>Support for custom labels and texts</li>
<li>Great control of every element in a figure</li>
<li>High-Quality Output in many formats</li>
<li>Very customizable in general</li>

```python
import matplotlib.pyplot as plt
#to avoid plt.show()
%matplotlib inline 
```

```python
import numpy as np
```

```python
x = np.arange(0,5)
y = np.arange(10,15)
```

```python
a = np.arange(40,50)
b = np.arange(50,60)
plt.plot(a,b)
```

```python
#plotting using matplotlib
plt.scatter(x,y, c='b')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Numeric Graph 2D')
# to save the output
plt.savefig('numTestGraph.png')
```

```python
# Plt Plot
y = x*x
plt.plot(x,y, 'r', marker='*', linestyle='--', linewidth = 2, markersize = 7)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('2-D Diagram')
```

```python
# Creating Subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'bo')
plt.subplot(2,2,3)
plt.plot(x,y,'g*--')
```

```python
x = np.arange(1,11)
y = 3 * x + 5
plt.title('Demo Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.plot(x,y)
plt.show()
```

```python
# Compute the x and y coordinates for points on a sine curve
x = np.arange(0,4 * np.pi, 0.1)
y = np.sin(x)
plt.title('Sin waveform')

#ploting the points
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,4 * np.pi, 0.1)
sin_y = np.sin(x)
cos_y = np.cos(x)

# Using subplot
#Plotting sin wave form
plt.subplot(2,1,1)
plt.plot(x, sin_y)
plt.title('Sin wave')

#Plotting cos waveform
plt.subplot(2,1,2)
plt.plot(x, cos_y , 'r--')
plt.title('Cos wave')

plt.show()
```

```python
# Bar plot

x = [2,5,12]
y = [11,23,12]

x2 = [8,7,9]
y2 = [2,4,6]

plt.bar(x,y)
plt.bar(x2,y2, color = 'r')
plt.title('Bar Graph')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

<font size='5'>Histogram</font>

```python
a = np.array([22,23,21,19,10,11,9,18,14,13,12,17,31])
plt.hist(a)
plt.title("Histogram")
plt.show()
```

<font size='5'>Box Plot</font>

```python
data = [np.random.normal(0, std, 100) for std in range(1,4)]

# rectangular box plot
plt.boxplot(data, vert = True, patch_artist=True)
```

<font size='5'>Pie Chart</font>

```python
# Data to plot

labels = 'Python', 'C++', 'JavaScript', 'Node.js'
sizes = [45, 35, 30, 40]
colors = ['red', 'gold', 'teal', 'green']
explode = (0.2, 0.1, 0.1, 0)

plt.pie(sizes, explode=explode, labels = labels, colors = colors, autopct='%1.1f%%', shadow=True)

plt.axis('equal')
plt.show()
```
