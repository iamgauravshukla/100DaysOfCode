```python
from io import StringIO , BytesIO
```

```python
data = ('col-1, col-2, col-3\n'
        'c,d,4\n'
        'a,b,1\n'
        'x,y,2\n'
        'odd,even,odd'
)
```

```python
pd.read_csv(StringIO(data))
```

```python
# read from specific column
df = pd.read_csv(StringIO(data), usecols=['col-1'])
```

```python
#specifying columns data types
data = (
        'a,b,c,d\n'
    '1,2,3,4\n'
    '5,6,7,8\n'
    '9,10,11,12'
)
```

```python
print(data)
```

```python
dataRead = pd.read_csv(StringIO(data),dtype=float)
```

```python
dataRead
```

```python
dataRead['a'][1]
```

```python
dataRead = pd.read_csv(StringIO(data), dtype={'b' : int, 'a' : float, 'c' : float, 'd' : object})
```

```python
dataRead['b']
```

```python
dataRead['a']
```

```python
# Index columns and trainig delimiters
```

```python
data = (' a,b,c\n'
       '1, apple, banana,3.2\n'
       '2, goat,cow,2.2')
```

```python
pd.read_csv(StringIO(data),index_col = 0)
```

```python
data2 = ('a,b,c\n'
       '2,apple, banana\n'
       '3,goat,cow')
```

```python
pd.read_csv(StringIO(data2), index_col=0)
```

```python
pd.read_csv(StringIO(data2), index_col=False)
```

```python
# Combining usecols and indexcols
data = ('a,b,c\n'
       '2,apple, banana\n'
       '3,goat,cow')
```

```python
pd.read_csv(StringIO(data), usecols=['b','c'], index_col=False)
```

```python
#Quoting and escape character
data = 'a,b\n"hello, \\"bob\\", how are you", 5'
```

```python
pd.read_csv(StringIO(data), escapechar='\\')
```
